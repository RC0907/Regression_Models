Importing libraries 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Importing Dataset where x  repressenting values of table from first row till last row,where y repressenting values of last column only.

```python
dataset = pd.read_csv('/Users/rahulchauhan/Documents/Machine Learning/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 8 - Decision Tree Regression/Python/Position_Salaries.csv')
x= dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

Importing The Descision Tree model on whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.plot(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
