
Importing Libraries 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Importing Dataset where x  repressenting values of table from first row till last row,where y repressenting values of last column only.

```python
dataset = pd.read_csv('/Users/rahulchauhan/Documents/Machine Learning/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv')
x= dataset.iloc[:,:-1].values
y = dataset.iloc[:,:-1].values
```

Importing train_test_split model on entire dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Importing Linear Regression Model on entire dataset

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(x_train,y_train)
```

```python
plt.plot(x_train,y_train,color = 'red')
plt.scatter(x_train,le.predict(x_train),color = 'blue')
plt.title('Salary Vs Experience')
plt.xlabel('Years of Experinece')
plt.ylabel('Salary')
```
